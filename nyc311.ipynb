### Droping useless columns 

```python
nyc311_copy.drop(['Bridge Highway Name','Bridge Highway Direction','Road Ramp','Facility Type','School or Citywide Complaint', 'Vehicle Type',
       'Taxi Company Borough', 'Taxi Pick Up Location','Bridge Highway Segment','Garage Lot Name','Ferry Direction','Ferry Terminal Name','Landmark','Intersection Street 1','Intersection Street 2'], axis = 1, inplace=True)
```

```python
import numpy as np
```

### Convert Created Date as time stamp

```python
create = pd.to_datetime(nyc311_copy['Created Date']).values
```

```python
create
```

### Convert Closed Date as time stamp

```python
closed = pd.to_datetime(nyc311_copy['Closed Date']).values
```

```python
closed
```

### Calculate Elapsed Time

```python
elapsed = closed - create
elapsed
```

### Converting into seconds

```python
elapsed = elapsed.astype('timedelta64[s]')
```

### Get the position of Closed Date in the dataframe

```python
closeddate_position = nyc311_copy.columns.get_loc("Closed Date")
closeddate_position
```

### inserting new columns create, close and elapsed in our nyc311_copy dataset

```python
nyc311_copy.insert(closeddate_position + 1, "Create", create)
```

```python
nyc311_copy.insert(closeddate_position + 2, "Close", closed)
```

```python
nyc311_copy.insert(closeddate_position + 3, "Elapsed", elapsed)
```

### Displaying Elapse time in week

```python
elapsed_time_counts_week = elapsed_time_counts.sort_index().head(7)
elapsed_time_counts_week
```

```python
elapsed_time_counts_week.plot.bar(elapsed_time_counts_week.index,elapsed_time_counts_week.values, figsize=(10,5))
plt.xlabel("Elapsed days")
plt.ylabel("Resolution count")
plt.title("Elapsed days analysis")
for i in range(elapsed_time_counts_week.size):
    plt.annotate(s = elapsed_time_counts_week[i], xy = (i,elapsed_time_counts_week[i] ), size = 10, color = 'darkgreen', weight = 'bold',ha='center')
plt.show()
```
